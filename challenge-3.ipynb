## Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

```python
# import required libraries
import numpy as np
import pandas as pd
```

Next, extract and import `Orders` dataset into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

```python
# your code here
nombre_archivo = "Orders.csv"
orders = pd.read_csv(nombre_archivo)
orders.head(10)
```

```python
orders.shape
```

```python
orders = orders.loc[:, ~orders.columns.str.contains('^Unnamed')]
orders.sample(10)
```

```python
print(orders.columns)
```

```python
total_amount_per_customer = orders.groupby('CustomerID')['amount_spent'].sum().reset_index()
```

```python
percentiles = total_amount_per_customer['amount_spent'].quantile([0.75, 0.95])


vip_customers = total_amount_per_customer[total_amount_per_customer['amount_spent'] > percentiles.loc[0.95]]
preferred_customers = total_amount_per_customer[(total_amount_per_customer['amount_spent'] >= percentiles.loc[0.75])&(total_amount_per_customer['amount_spent'] <= percentiles.loc[0.95])]
```

## Q2: How to identify which country has the most VIP Customers?

```python
vip_customers_with_country = pd.merge(orders, vip_customers, on='CustomerID', how='inner')
vip_customers_by_country = vip_customers_with_country.groupby('Country')['CustomerID'].count().reset_index()


country_with_most_vip_customers = vip_customers_by_country.loc[vip_customers_by_country['CustomerID'].idxmax()]
country_with_most_vip_customers["Country"]
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
vip_and_preferred_customers = pd.merge(orders, vip_customers, on='CustomerID', how='inner')
vip_and_preferred_customers = pd.merge(vip_and_preferred_customers, preferred_customers, on='CustomerID', how='outer')

vip_and_preferred_by_country = vip_and_preferred_customers.groupby('Country')['CustomerID'].count().reset_index()

country_with_most_vip_and_preferred_customers = vip_and_preferred_by_country.loc[vip_and_preferred_by_country['CustomerID'].idxmax()]
country_with_most_vip_and_preferred_customers["Country"]
```
